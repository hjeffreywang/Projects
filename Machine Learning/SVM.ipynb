# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set and remove features. This will also be used as your x values.
target = df["koi_disposition"]
data = df.drop(["koi_disposition","koi_fpflag_nt"], axis=1)
features = data.columns
data.head(5)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=0)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data

from sklearn.preprocessing import MinMaxScaler
X_minmax = MinMaxScaler().fit(X_train)

X_train = X_minmax.transform(X_train)
X_test = X_minmax.transform(X_test)
```

# Train the Model



```python
from sklearn.svm import SVC 

SVC = SVC(kernel='poly')
SVC.fit(X_train, y_train)
```

```python

print(f"Training Data Score: {SVC.score(X_train, y_train)}")
print(f"Testing Data Score: {SVC.score(X_test, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
# Train the model with GridSearch
from sklearn.model_selection import GridSearchCV
parameters = {'C': [1, 5, 10],
              'gamma': [0.0001, 0.001, 0.005]}
grid = GridSearchCV(SVC, parameters, verbose=3, n_jobs=-1)
#n_jobs for faster processing
```

```python

# Train the model with GridSearch
grid.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {grid.best_params_}")
print(f"Training Data Score: {grid.best_score_}")
```
